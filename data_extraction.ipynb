```python
import pathlib
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
DATA_DIR = pathlib.Path.cwd() / 'data'
# Trending videos from Canada youtube
CA_videos = pd.read_csv(DATA_DIR / 'CAvideos.csv')
print(DATA_DIR)
```

```python
CA_videos.info()
```

```python
# Filtrando videos que nao tem caracteres em ascii
# Tamanho do dataset filtrado: 33484
CA_videos = CA_videos[CA_videos['title'].map(lambda x: x.isascii())]
CA_videos = CA_videos[CA_videos['video_error_or_removed'] == False]
```

```python
CA_videos.head()
```

```python
# Inicialmente irei dividir o dataset em blocos com 5000 linhas
size = len(CA_videos)
# CA1 = CA_videos[0:size//6] FEITO
CA2 = CA_videos[size//6-1:size//3]
CA3 = CA_videos[size//3-1:size//2]
CA4 = CA_videos[size//2-1:2*size//3]
CA5 = CA_videos[2*size-1//3:5*size//6]
CA5 = CA_videos[5*size-1//6:size]
```

```python
def get_subtitles(x):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(x)
        text = ""
        for subtitle in transcript:
            text+=" " + subtitle['text']
        print(f"Foi para id : {x}")
        return text
    except:
        print("Video nao possui subtitles")
        return None
    
CA2['subtitles'] = CA2['video_id'].apply(get_subtitles)
```
