```python
import pathlib
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
DATA_DIR = pathlib.Path.cwd() / 'data'
# Trending videos from Canada youtube
US_videos = pd.read_csv(DATA_DIR / 'USvideos.csv')
print(DATA_DIR)
```

```python
US_videos.info()
```

```python
# Filtrando videos que nao tem caracteres em ascii
# Tamanho do dataset filtrado: 33484
US_videos = US_videos[US_videos['title'].map(lambda x: x.isascii())]
US_videos = US_videos[US_videos['video_error_or_removed'] == False]
```

```python
US_videos.head()
```

```python
# Inicialmente irei dividir o dataset em blocos com 5000 linhas
size = len(US_videos)
US1 = US_videos[0:size//3] 
US2 = US_videos[size//3:2*size//3] 
US3 = US_videos[2*size//3:size] 
```

```python
def get_subtitles(x):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(x)
        text = ""
        for subtitle in transcript:
            text+=" " + subtitle['text']
        print(f"Foi para id : {x}")
        return text
    except:
        print("Video nao possui subtitles")
        return None
    
US3['subtitles'] = US3['video_id'].apply(get_subtitles)
```

```python
US3_final = US3.copy()
```

```python
US3_final = US3_final[US3_final['subtitles'].notna()]
```

```python
US3_final.info()
```

```python
US3_final.to_csv(DATA_DIR / 'US_videos' / 'US3.csv')
```
